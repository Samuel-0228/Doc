```python
docs = {
    "doc1": "Addis Ababa University students are working on projects.",
    "doc2": "Coffee plays a major role in Ethiopian culture.",
    "doc3": "Addis Ababa is the capital city of Ethiopia.",
    "doc4": "Students study hard at the university every day."
}
def search(query):
    results = []
    for doc_id, text in docs.items():
        if query.lower() in text.lower():  # check if query word is inside the text
            results.append(doc_id)
    return results
print("Search for 'Addis':", search("Addis"))
print("Search for 'Coffee':", search("Coffee"))
```

```python
import re
from collections import defaultdict

# -----------------------------
# Step 1: Sample English Documents
# -----------------------------
docs = {
    "doc1": "Addis Ababa University students are working on projects.",
    "doc2": "Coffee plays a major role in Ethiopian culture.",
    "doc3": "Addis Ababa is the capital city of Ethiopia.",
    "doc4": "Students study hard at the university every day."
}

# -----------------------------
# Step 2: Tokenization
# -----------------------------
def tokenize(text):
    # Use regex to split text into words (lowercased, no punctuation)
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

tokenized_docs = {doc_id: tokenize(text) for doc_id, text in docs.items()}

# -----------------------------
# Step 3: Stop-word Removal
# -----------------------------
stop_words = {"is", "the", "a", "on", "in", "at", "are", "of", "every"}

def remove_stopwords(tokens):
    # Filter out common stop-words
    return [t for t in tokens if t not in stop_words]

cleaned_docs = {doc_id: remove_stopwords(tokens) for doc_id, tokens in tokenized_docs.items()}

# -----------------------------
# Step 4: Build Inverted Index
# -----------------------------
index = defaultdict(list)
for doc_id, tokens in cleaned_docs.items():
    for token in tokens:
        if doc_id not in index[token]:
            index[token].append(doc_id)

print("=== Inverted Index ===")
for term, postings in index.items():
    print(term, ":", postings)

# -----------------------------
# Step 5: Boolean Retrieval
# -----------------------------
def boolean_and(term1, term2):
    return set(index.get(term1, [])) & set(index.get(term2, []))

def boolean_or(term1, term2):
    return set(index.get(term1, [])) | set(index.get(term2, []))

def boolean_not(term):
    return set(docs.keys()) - set(index.get(term, []))

print("\n=== Boolean Retrieval Demo ===")
print("AND (addis, ababa):", boolean_and("addis", "ababa"))
print("OR (coffee, students):", boolean_or("coffee", "students"))
print("NOT (university):", boolean_not("university"))
```
