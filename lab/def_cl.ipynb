```python
def main():
    print("Hello, World!")
    name = "calculater"
    
    while True:
        print("""
                =====Menu=====
                1. Addition
                2. Multiply  
                3. Division
                4. Subtract
                5. Exit
                Enter choice: """)
        
        choice = int(input())
        
        if choice == 5:
            break
            
        print("Enter number 1: ", end="")
        num = int(input())
        
        print("Enter number 2: ", end="")
        num2 = int(input())
        
        if choice == 1:
            print(num + num2)
        elif choice == 2:
            print(num * num2)
        elif choice == 3:
            print(num / num2)
        elif choice == 4:
            print(num - num2)
        elif choice == 5:
            print("Exit")
        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
